## identifying&handling outliers
**Outliers**-data points that significantly differ from the rest of the dataset
in this step we to spot what are the features that do not add value to the continuos data.we do this by identifying outliers by using statistical methods in this case **{Interquartile range}**
**Treatment** we decide whether to remove, cap or transform outliers


```python
#identifying outliers in our continuos data
Q1 = df['Total.Fatal.Injuries'].quantile(0.25)
Q3 = df['Total.Fatal.Injuries'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Total.Fatal.Injuries'] < (Q1 - 1.5 * IQR)) | (df['Total.Fatal.Injuries'] > (Q3 + 1.5 * IQR))]
outliers
```

```python
Q1 = df['Number.of.Engines'].quantile(0.25)
Q3 = df['Number.of.Engines'].quantile(0.75)
IQR = Q3 - Q1
outliers_1 = df[(df['Number.of.Engines'] < (Q1 - 1.5 * IQR)) | (df['Number.of.Engines'] > (Q3 + 1.5 * IQR))]
outliers_1
```

## Handling missing values 
-some of the data present might have missing values 
-it is upto us to find the best mode of **imputation-missing data is filled in to create a complete data matrix that can be analysed using standard methods**
-imputation can be done through-**mean,median,mode**
-This is done so as to enable us keep our original data & key points to calculate keen insights that help in focusing on our objective

```python
# filling missing values of continuos data
relevant_columns = ["Number.of.Engines", "Total.Fatal.Injuries", "Total.Serious.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]
relevant_columns
```

```python
# assigned list of columns and re_assigning missing values with their median
df[relevant_columns] = df[relevant_columns].fillna(df[relevant_columns].median())
```
