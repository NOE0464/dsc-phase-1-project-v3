## identifying&handling outliers
**Outliers**-data points that significantly differ from the rest of the dataset
in this step we to spot what are the features that do not add value to the continuos data.we do this by identifying outliers by using statistical methods in this case **{Interquartile range}**
**Treatment** we decide whether to remove, cap or transform outliers


```python
#identifying outliers in our continuos data
# outliers in Total fatal injuries
Q1 = df['Total.Fatal.Injuries'].quantile(0.25)
Q3 = df['Total.Fatal.Injuries'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Total.Fatal.Injuries'] < (Q1 - 1.5 * IQR)) | (df['Total.Fatal.Injuries'] > (Q3 + 1.5 * IQR))]
outliers
```

```python
# outliers in Number of Engines
Q1 = df['Number.of.Engines'].quantile(0.25)
Q3 = df['Number.of.Engines'].quantile(0.75)
IQR = Q3 - Q1
outliers_1 = df[(df['Number.of.Engines'] < (Q1 - 1.5 * IQR)) | (df['Number.of.Engines'] > (Q3 + 1.5 * IQR))]
outliers_1
```

```python
# outliers in Total serious injuries
Q1 = df['Total.Serious.Injuries'].quantile(0.25)
Q3 = df['Total.Serious.Injuries'].quantile(0.75)
IQR = Q3 - Q1
outliers_2 = df[(df['Total.Serious.Injuries'] < (Q1 - 1.5 * IQR)) | (df['Total.Serious.Injuries'] > (Q3 + 1.5 * IQR))]
outliers_2
```

```python
# outliers in Total minor Injuries
Q1 = df['Total.Minor.Injuries'].quantile(0.25)
Q3 = df['Total.Minor.Injuries'].quantile(0.75)
IQR = Q3 - Q1
outliers_3 = df[(df['Total.Minor.Injuries'] < (Q1 - 1.5 * IQR)) | (df['Total.Minor.Injuries'] > (Q3 + 1.5 * IQR))]
outliers_3
```

```python
Q1 = df['Total.Uninjured'].quantile(0.25)
Q3 = df['Total.Uninjured'].quantile(0.75)
IQR = Q3 - Q1
outliers_4 = df[(df['Total.Uninjured'] < (Q1 - 1.5 * IQR)) | (df['Total.Uninjured'] > (Q3 + 1.5 * IQR))]
outliers_4
```

## Handling missing values 
-some of the data present might have missing values 
-it is upto us to find the best mode of **imputation-missing data is filled in to create a complete data matrix that can be analysed using standard methods**
-imputation can be done through-**mean,median,mode**
-This is done so as to enable us keep our original data & key points to calculate keen insights that help in focusing on our objective

```python
# filling missing values of continuos data
relevant_columns = ["Number.of.Engines", "Total.Fatal.Injuries", "Total.Serious.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]
relevant_columns
```

```python
# assigned list of columns and re_assigning missing values with their median
df[relevant_columns] = df[relevant_columns].fillna(df[relevant_columns].median())
```

```python
#Forms a sturdy basis of our project goals;category={type,kind,classification,group,class}of aircraft is safest
# Group by Aircraft Category and calculate mean number of accidents
Viz = df.groupby('Aircraft.Category').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
Viz
```

```python
# decision on most relevant aircraft to start business
flight = df[df['Aircraft.Category'] == 'Airplane'].agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
flight
```

```python
# #Group by Country to gauge competition and calculate mean number of accidents
VIZ_1 = df.groupby('Country').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
VIZ_1
```

```python
#Group by Amateur built  and calculate mean number of accidents
VIZ_4 = df.groupby('Amateur.Built').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
VIZ_4
```

#####  stacked bar plot
for showing hierarchical data :Visualize and compare subcategory data: Stacked bar plots allow you to break down a categorical variable into subcategories, enabling easy comparison of their contributions to the overall total.

```python
# Data
countries = VIZ_1['Country']
fatal_injuries = VIZ_1['Total.Fatal.Injuries']
uninjured = VIZ_1['Total.Uninjured']
# Plot
fig, ax = plt.subplots(figsize=(10, 6))
index = np.arange(len(countries))
bar_width = 0.5
# Create stacked bar plot
ax.bar(index, fatal_injuries, bar_width, label='Fatal Injuries', color='red')
ax.bar(index, uninjured, bar_width, bottom=fatal_injuries, label='Uninjured', color='green')

# Adding labels, title, and legend
ax.set_xlabel('Country')
ax.set_ylabel('Number of People')
ax.set_title('Fatality by Country')
ax.set_xticks(index)
ax.set_xticklabels(countries, rotation=90)
ax.legend()
plt.tight_layout()
plt,save
```

##### Countplots
-for comparing categories
-visual appeal and handling large datasetd

```python
# Get the top 10 airplane makes by accident count
risky_makes = df['Make'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the top 10 makes
risks_10_df = df[df['Make'].isin(risky_makes)]

# Create the plot
plt.figure(figsize=(10, 8))
sns.countplot(data=risks_10_df, x='Make', order=risky_makes)
plt.title('Top 10 Airplane Makes & Overall accidents')
plt.xticks(rotation=10, ha='right')  # Rotate x labels for better readability
plt.show()
```

```python
#Get the top 10 Purpose of Flight by accident count
reason = df['Purpose.of.flight'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the top 10 makes
why = df[df['Purpose.of.flight'].isin(reason)]

# Create the plot
plt.figure(figsize=(10, 8))
sns.countplot(data=why, x='Purpose.of.flight', order=reason)
plt.title('Flight reasons and the Number of Accidents')
plt.xticks(rotation=30, ha='right')  # Rotate x labels for better readability
plt.show()
```
